```python
# Veri artırma işlevleri tanımlanıyor.
def add_noise(data, x):
    # Veriye rastgele bir gürültü ekleniyor.
    noise = np.random.randn(len(data))
    data_noise = data + x * noise
    return data_noise

def shift(data, x):
    # Veriler belirli bir miktarda kaydırılıyor.
    return np.roll(data, x)

def stretch(data, rate):
    # Zaman ekseninde veriler uzatılıyor veya sıkıştırılıyor.
    data = librosa.effects.time_stretch(data, rate=1.0)
    return data

def pitch_shift(data, rate):
    # Verilerdeki frekans kaydırılarak pitch değişimi uygulanıyor.
    data = librosa.effects.pitch_shift(data, sr=220250, n_steps=rate)
    return data

```

```python
# Bu fonksiyon, bir ses dosyasını yükler, veri artırma teknikleri uygular ve her işlem sonrası ses özelliklerini görselleştirir.
def show_audio_features(audio_data_path):
    # Ses dosyasını yükleme
    # `y`: Ses sinyali (dizi şeklinde)
    # `sr`: Örnekleme oranı (standart olarak 22050 Hz)
    y, sr = librosa.load(audio_data_path)
    
    # Veri artırma işlemleri
    # Gürültü ekleniyor
    y_noise = add_noise(y, 0.0008)
    # Sinyal kaydırılıyor
    y_shift = shift(y, 3200)
    # Sinyal zaman ekseninde uzatılıyor
    y_stretch_1 = stretch(y, 1.2)
    # Sinyal zaman ekseninde sıkıştırılıyor
    y_stretch_2 = stretch(y, 0.8)
    # Sinyalin pitch'i (frekansı) değiştirilerek ton kaydırma uygulanıyor
    y_pitch_shift = pitch_shift(y, 3)
    
    # MFCC (Mel Frequency Cepstral Coefficients) özellikleri çıkarılıyor
    # Bu özellikler sesin frekans bazlı bir temsili olarak kullanılır
    y = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=50)
    y_noise = librosa.feature.mfcc(y=y_noise, sr=sr, n_mfcc=50)
    y_shift = librosa.feature.mfcc(y=y_shift, sr=sr, n_mfcc=50)
    y_stretch_1 = librosa.feature.mfcc(y=y_stretch_1, sr=sr, n_mfcc=50)
    y_stretch_2 = librosa.feature.mfcc(y=y_stretch_2, sr=sr, n_mfcc=50)
    y_pitch_shift = librosa.feature.mfcc(y=y_pitch_shift, sr=sr, n_mfcc=50)
    
    # Görselleştirme işlemleri
    # Grafik penceresi boyutu ayarlanıyor
    plt.figure(figsize=(20, 8))
    
    # Orijinal sesin MFCC temsili
    plt.subplot(3, 2, 1)
    librosa.display.specshow(librosa.power_to_db(y, ref=np.max),
                             y_axis='mel',  # Frekans ekseni mel ölçeğinde gösteriliyor
                             fmax=8000,     # Maksimum frekans 8000 Hz olarak sınırlandırılıyor
                             x_axis='time') # Zaman ekseni
    plt.colorbar(format='%+2.0f dB')  # Renk çubuğu ekleniyor
    plt.title('Original')  # Grafik başlığı

    # Gürültü eklenmiş sesin MFCC temsili
    plt.subplot(3, 2, 2)
    librosa.display.specshow(librosa.power_to_db(y_noise, ref=np.max),
                             y_axis='mel', fmax=8000, x_axis='time')
    plt.colorbar(format='%+2.0f dB')
    plt.title('Noise')

    # Kaydırılmış sesin MFCC temsili
    plt.subplot(3, 2, 3)
    librosa.display.specshow(librosa.power_to_db(y_shift, ref=np.max),
                             y_axis='mel', fmax=8000, x_axis='time')
    plt.colorbar(format='%+2.0f dB')
    plt.title('Shift')
    
    # Uzatılmış sesin MFCC temsili
    plt.subplot(3, 2, 4)
    librosa.display.specshow(librosa.power_to_db(y_stretch_1, ref=np.max),
                             y_axis='mel', fmax=8000, x_axis='time')
    plt.colorbar(format='%+2.0f dB')
    plt.title('Stretch 1')
    
    # Sıkıştırılmış sesin MFCC temsili
    plt.subplot(3, 2, 5)
    librosa.display.specshow(librosa.power_to_db(y_stretch_2, ref=np.max),
                             y_axis='mel', fmax=8000, x_axis='time')
    plt.colorbar(format='%+2.0f dB')
    plt.title('Stretch 2')
    
    # Pitch kaydırılmış sesin MFCC temsili
    plt.subplot(3, 2, 6)
    librosa.display.specshow(librosa.power_to_db(y_pitch_shift, ref=np.max),
                             y_axis='mel', fmax=8000, x_axis='time')
    plt.colorbar(format='%+2.0f dB')
    plt.title('Pitch Shift')
    
    # Tüm grafikleri düzenlemek için sıkı bir düzen yerleşimi uygulanıyor
    plt.tight_layout()
```

```python
def mfccs_feature_exteraction(dir_):
    '''
        Ses verilerinden MFCC (Mel-Frekans Kepstral Katsayıları) özelliklerini çıkarır, veri artırımı yapar ve veri dengesizliğini gidermek için belirli hasta ID'lerini hariç tutar.
        Parametreler:
            dir_: Ses dosyalarının bulunduğu dizin.
        Dönen Değerler:
            X_data: Çıkarılan özelliklerin dizisi.
            y_data: Özelliklere karşılık gelen etiketlerin dizisi.
    '''
    X_ = []  # Ses dosyalarından çıkarılan özelliklerin saklandığı liste.
    y_ = []  # Ses dosyalarına karşılık gelen etiketlerin saklandığı liste.

    COPD = []  # COPD hastalarına ait ses dosyalarını takip eden liste.
    copd_count = 0  # Belirli bir COPD hastasına ait işlenen ses dosyalarının sayısını tutar.

    data = diagnosis_df  # Hasta bilgileri ve teşhislerini içeren DataFrame.
    features = 52  # Çıkarılacak MFCC özelliklerinin sayısı.

    # Belirtilen dizindeki dosyalar üzerinde döngü.
    for soundDir in (os.listdir(dir_)):
        # Sadece '.wav' uzantılı dosyaları ve belirli hasta ID'lerini hariç tutarak işleme devam edilir.
        if soundDir[-3:] == 'wav' and soundDir[:3] != '103' and soundDir[:3] != '108' and soundDir[:3] != '115':

            # Hasta ID'sine göre teşhisi alınır.
            p = list(data[data['patient_id'] == int(soundDir[:3])]['disease'])[0]

            # Eğer teşhis COPD ise
            if p == 'COPD':
                if (soundDir[:6] in COPD) and copd_count < 2:
                    # Ses verisi yüklenir ve MFCC özellikleri çıkarılır.
                    data_x, sampling_rate = librosa.load(dir_ + soundDir, res_type='kaiser_fast')
                    mfccs = np.mean(librosa.feature.mfcc(y=data_x, sr=sampling_rate, n_mfcc=features).T, axis=0)
                    COPD.append(soundDir[:6])
                    copd_count += 1
                    X_.append(mfccs)
                    y_.append(list(data[data['patient_id'] == int(soundDir[:3])]['disease'])[0])
                    
                if soundDir[:6] not in COPD:
                    # Yeni bir COPD hastası için ses verisi işlenir.
                    data_x, sampling_rate = librosa.load(dir_ + soundDir, res_type='kaiser_fast')
                    mfccs = np.mean(librosa.feature.mfcc(y=data_x, sr=sampling_rate, n_mfcc=features).T, axis=0)
                    COPD.append(soundDir[:6])
                    copd_count = 0
                    X_.append(mfccs)
                    y_.append(list(data[data['patient_id'] == int(soundDir[:3])]['disease'])[0])
                
            # Eğer teşhis COPD değilse
            if p != 'COPD':
                # Teşhis Bronşektazi veya Bronşiolit ise
                if p == 'Bronchiectasis' or p == 'Bronchiolitis':
                    data_x, sampling_rate = librosa.load(dir_ + soundDir, res_type='kaiser_fast')
                    mfccs = np.mean(librosa.feature.mfcc(y=data_x, sr=sampling_rate, n_mfcc=features).T, axis=0)
                    X_.append(mfccs)
                    y_.append('Bronchiolitis')

                    # Gürültü eklenmiş veri ile özellik çıkarımı.
                    data_noise = add_noise(data_x, 0.001)
                    mfccs_noise = np.mean(librosa.feature.mfcc(y=data_noise, sr=sampling_rate, n_mfcc=features).T, axis=0)
                    X_.append(mfccs_noise)
                    y_.append('Bronchiolitis')

                    # Kaydırılmış veri ile özellik çıkarımı.
                    data_shift = shift(data_x, 1600)
                    mfccs_shift = np.mean(librosa.feature.mfcc(y=data_shift, sr=sampling_rate, n_mfcc=features).T, axis=0)
                    X_.append(mfccs_shift)
                    y_.append('Bronchiolitis')

                    # Uzatılmış (stretch) veri ile özellik çıkarımı.
                    data_stretch = stretch(data_x, 1.2)
                    mfccs_stretch = np.mean(librosa.feature.mfcc(y=data_stretch, sr=sampling_rate, n_mfcc=features).T, axis=0)
                    X_.append(mfccs_stretch)
                    y_.append('Bronchiolitis')

                    # Sıkıştırılmış (stretch) veri ile özellik çıkarımı.
                    data_stretch_2 = stretch(data_x, 0.8)
                    mfccs_stretch_2 = np.mean(librosa.feature.mfcc(y=data_stretch_2, sr=sampling_rate, n_mfcc=features).T, axis=0)
                    X_.append(mfccs_stretch_2)
                    y_.append('Bronchiolitis')

                    # Pitch (ses perdesi) değiştirilmiş veri ile özellik çıkarımı.
                    data_pitch_shift = pitch_shift(data_x, 3)
                    mfccs_stretch = np.mean(librosa.feature.melspectrogram(y=data_pitch_shift, sr=sampling_rate, n_mels=features).T, axis=0)
                    X_.append(mfccs_stretch)
                    y_.append('Bronchiolitis')

                else:
                    # Diğer teşhisler için aynı işlem yapılır.
                    data_x, sampling_rate = librosa.load(dir_ + soundDir, res_type='kaiser_fast')
                    mfccs = np.mean(librosa.feature.mfcc(y=data_x, sr=sampling_rate, n_mfcc=features).T, axis=0)
                    X_.append(mfccs)
                    y_.append(list(data[data['patient_id'] == int(soundDir[:3])]['disease'])[0])

                    # Gürültü, kaydırma, uzatma ve pitch değişikliği için veri artırımı yapılır.
                    data_noise = add_noise(data_x, 0.001)
                    mfccs_noise = np.mean(librosa.feature.mfcc(y=data_noise, sr=sampling_rate, n_mfcc=features).T, axis=0)
                    X_.append(mfccs_noise)
                    y_.append(p)

                    data_shift = shift(data_x, 1600)
                    mfccs_shift = np.mean(librosa.feature.mfcc(y=data_shift, sr=sampling_rate, n_mfcc=features).T, axis=0)
                    X_.append(mfccs_shift)
                    y_.append(p)

                    data_stretch = stretch(data_x, 1.2)
                    mfccs_stretch = np.mean(librosa.feature.mfcc(y=data_stretch, sr=sampling_rate, n_mfcc=features).T, axis=0)
                    X_.append(mfccs_stretch)
                    y_.append(p)

                    data_stretch_2 = stretch(data_x, 0.8)
                    mfccs_stretch_2 = np.mean(librosa.feature.mfcc(y=data_stretch_2, sr=sampling_rate, n_mfcc=features).T, axis=0)
                    X_.append(mfccs_stretch_2)
                    y_.append(p)

                    data_pitch_shift = pitch_shift(data_x, 3)
                    mfccs_stretch = np.mean(librosa.feature.melspectrogram(y=data_pitch_shift, sr=sampling_rate, n_mels=features).T, axis=0)
                    X_.append(mfccs_stretch)
                    y_.append(p)

    # Özellikler ve etiketler numpy dizisine dönüştürülür.
    X_data = np.array(X_)
    y_data = np.array(y_)

    return X_data, y_data
```

```python
# mfccs_features ve lables dizilerini, eğitim, doğrulama ve test verilerine ayırıyoruz.
# train_test_split fonksiyonu, veriyi belirtilen oranlarda (test_size) eğitim ve test (doğrulama) verilerine böler.
# random_state, her seferinde aynı rastgele veriyi elde etmek için sabit bir değer sağlar.
# Burada test verisi olarak %17.5 oranı belirlenmiştir.

mfcc_train, mfcc_val, lables_train, lables_val = train_test_split(mfccs_features, lables, test_size=0.175, random_state=10)

# Eğitim verisini daha küçük bir test (test_size %7.5) verisine ayırıyoruz.
# Bu sayede nihai test veri kümesi elde edilir.
mfcc_train, mfcc_test, lables_train, lables_test = train_test_split(mfcc_train, lables_train, test_size=0.075, random_state=10)

# mfcc_train, mfcc_val ve mfcc_test dizilerinin boyutlarını yazdırıyoruz.
# Bu, verilerin hangi oranlarda bölündüğünü ve her bir kümenin boyutunu görmemizi sağlar.
print(mfcc_train.shape, mfcc_val.shape, mfcc_test.shape)

# lables_train, lables_val ve lables_test dizilerinin boyutlarını yazdırıyoruz.
# Bu, etiketlerin (lables) eğitim, doğrulama ve test kümelerine nasıl dağıldığını gösterir.
print(lables_train.shape, lables_val.shape, lables_test.shape)
```

```python
# Bu fonksiyon, eğitim sürecinde kaydedilen kayıp (loss) ve doğruluk (accuracy) değerlerinin grafiklerini çizer.
# "history" nesnesi, modelin eğitim sürecindeki kayıp ve doğruluk değerlerini içerir.

def plot_loss_curves(history):
    # Eğitim kaybı (loss) ve doğrulama kaybı (val_loss) değerlerini history nesnesinden alıyoruz.
    loss = history.history['loss']
    val_loss = history.history['val_loss']

    # Eğitim doğruluğu (accuracy) ve doğrulama doğruluğu (val_accuracy) değerlerini alıyoruz.
    accuracy = history.history['accuracy']
    val_accuracy = history.history['val_accuracy']

    # Epoch sayısını belirliyoruz. Bu, kayıp ve doğruluk değerlerinin kaç epoch (dönem) boyunca ölçüldüğünü gösterir.
    epochs = range(len(history.history['loss']))

    # Doğruluk (accuracy) grafiğini çiziyoruz.
    # Eğitim doğruluğunu ve doğrulama doğruluğunu aynı grafikte görselleştiriyoruz.
    plt.figure()  # Yeni bir figure oluşturuyoruz
    plt.grid()  # Grafik üzerinde ızgara (grid) gösteriyoruz
    plt.plot(epochs, accuracy, label='training_accuracy')  # Eğitim doğruluğunu çiziyoruz
    plt.plot(epochs, val_accuracy, label='val_accuracy')  # Doğrulama doğruluğunu çiziyoruz
    plt.title('Accuracy')  # Grafik başlığını ekliyoruz
    plt.xlabel('Epochs')  # X eksenine 'Epochs' etiketini koyuyoruz
    plt.legend()  # Grafik üzerinde doğruluklar için etiketleri gösteriyoruz
```

```python
# Bu kod parçacığı, eğitim, doğrulama ve test veri setlerindeki verileri, GRU (Gated Recurrent Unit) gibi RNN tabanlı bir modelin girişine uygun hale getirmek için genişletir.

# mfcc_train, mfcc_val ve mfcc_test veri setlerini 3D hale getiriyoruz (batch_size, timesteps, features).
# axis=1 parametresi, her bir örneği tek bir zaman adımı (timesteps) gibi kabul ederek, verileri bir boyut daha genişletiyor.

x_train_gru = np.expand_dims(mfcc_train, axis=1)  # Eğitim verisini 3D hale getiriyoruz (batch_size, 1, features)
x_val_gru = np.expand_dims(mfcc_val, axis=1)  # Doğrulama verisini 3D hale getiriyoruz (batch_size, 1, features)
x_test_gru = np.expand_dims(mfcc_test, axis=1)  # Test verisini 3D hale getiriyoruz (batch_size, 1, features)

# Etiket verilerini de 2D hale getiriyoruz. GRU modeli için, etiketlerin her biri de bir zaman adımı olarak kabul edilebilir.
y_train_gru = np.expand_dims(lables_train, axis=1)  # Eğitim etiketlerini 2D hale getiriyoruz (batch_size, 1)
y_val_gru = np.expand_dims(lables_val, axis=1)  # Doğrulama etiketlerini 2D hale getiriyoruz (batch_size, 1)
y_test_gru = np.expand_dims(lables_test, axis=1)  # Test etiketlerini 2D hale getiriyoruz (batch_size, 1)
```

```python
# Giriş katmanı (input layer) - (batch_size, 1, 52) boyutunda
Input_Sample = Input(shape=(1,52))

# 1D Konvolüsyon katmanları - Özellik çıkarımı için kullanılır
model_conv = Conv1D(256, kernel_size=5, strides=1, padding='same', activation='relu')(Input_Sample)  # 1D Conv katmanı
model_conv = MaxPooling1D(pool_size=2, strides=2, padding='same')(model_conv)  # Max Pooling
model_conv = BatchNormalization()(model_conv)  # Normalizasyon

model_conv = Conv1D(512, kernel_size=5, strides=1, padding='same', activation='relu')(model_conv)  # 1D Conv katmanı
model_conv = MaxPooling1D(pool_size=2, strides=2, padding='same')(model_conv)  # Max Pooling
model_conv = BatchNormalization()(model_conv)  # Normalizasyon

# GRU katmanları - Zaman serisi verileri için geri beslemeli ağlar
model_2_1 = GRU(32, return_sequences=True, activation='tanh', go_backwards=True)(model_conv)  # GRU katmanı
model_2 = GRU(128, return_sequences=True, activation='tanh', go_backwards=True)(model_2_1)  # GRU katmanı

model_3 = GRU(64, return_sequences=True, activation='tanh', go_backwards=True)(model_conv)  # GRU katmanı
model_3 = GRU(128, return_sequences=True, activation='tanh', go_backwards=True)(model_3)  # GRU katmanı

model_x = GRU(64, return_sequences=True, activation='tanh', go_backwards=True)(model_conv)  # GRU katmanı
model_x = GRU(128, return_sequences=True, activation='tanh', go_backwards=True)(model_x)  # GRU katmanı

# Katmanların birleştirilmesi (concatenation)
model_add_1 = add([model_3, model_2, model_x])  # Üç farklı GRU modelini toplama

# Daha fazla GRU katmanları ekleniyor
model_5 = GRU(128, return_sequences=True, activation='tanh', go_backwards=True)(model_add_1)  # GRU katmanı
model_5 = GRU(32, return_sequences=True, activation='tanh', go_backwards=True)(model_5)  # GRU katmanı

model_6 = GRU(64, return_sequences=True, activation='tanh', go_backwards=True)(model_add_1)  # GRU katmanı
model_6 = GRU(32, return_sequences=True, activation='tanh', go_backwards=True)(model_6)  # GRU katmanı

# Katmanların yeniden birleştirilmesi
model_add_2 = add([model_5, model_6, model_2_1])

# Yoğun (Dense) katmanlar ve Leaky ReLU aktivasyonu ile özellik çıkarmaya devam
model_7 = Dense(32, activation=None)(model_add_2)  # Dense katman
model_7 = LeakyReLU()(model_7)  # Leaky ReLU aktivasyonu
model_7 = Dense(128, activation=None)(model_7)  # Dense katman
model_7 = LeakyReLU()(model_7)  # Leaky ReLU aktivasyonu

model_9 = Dense(64, activation=None)(model_add_2)  # Dense katman
model_9 = LeakyReLU()(model_9)  # Leaky ReLU aktivasyonu
model_9 = Dense(128, activation=None)(model_9)  # Dense katman
model_9 = LeakyReLU()(model_9)  # Leaky ReLU aktivasyonu

# Katmanların birleştirilmesi
model_add_3 = add([model_7, model_9])

# Son düz katmanlar
model_10 = Dense(64, activation=None)(model_add_3)  # Dense katman
model_10 = LeakyReLU()(model_10)  # Leaky ReLU aktivasyonu

model_10 = Dense(32, activation=None)(model_10)  # Dense katman
model_10 = LeakyReLU()(model_10)  # Leaky ReLU aktivasyonu

# Çıktı katmanı (Softmax ile çok sınıflı sınıflandırma)
model_10 = Dense(5, activation="softmax")(model_10)  # 5 sınıf için softmax çıkışı

# Modelin tanımlanması
gru_model = Model(inputs=Input_Sample, outputs=model_10)

# Modelin özetini yazdırıyoruz
gru_model.summary()
```

```python
# Adam optimizer'ını tanımlıyoruz. Bu optimizer, her parametre için farklı öğrenme oranları kullanarak daha hızlı ve etkili bir gradyan inişi gerçekleştirir.
# Burada öğrenme oranı (learning rate) 0.0001 olarak belirlenmiş, bu oldukça küçük bir öğrenme oranı olup, modelin daha dikkatli öğrenmesini sağlar.
optimiser = tf.keras.optimizers.Adam(learning_rate = 0.0001)

# Modeli derliyoruz. 
# optimizer: Adam optimizer'ı kullanılıyor.
# loss: Modelin kayıp fonksiyonu olarak 'categorical_crossentropy' seçilmiş. Bu, çok sınıflı sınıflandırma problemleri için yaygın olarak kullanılan bir kayıp fonksiyonudur.
# metrics: Modelin doğruluğunu (accuracy) takip ediyoruz.
gru_model.compile(optimizer=optimiser, loss='categorical_crossentropy', metrics=['accuracy'])

# Callback fonksiyonlarını tanımlıyoruz.
cb = [
    # EarlyStopping: Modelin doğruluğu belirli bir seviyeye ulaştığında eğitim durdurulur.
    # patience=300: Eğer doğruluk 300 epoch boyunca artmazsa eğitim durdurulur.
    # monitor='accuracy': Doğruluk değerine bakılarak erken durdurma yapılır.
    # mode='max': Doğruluğun maksimum değere ulaşması beklenir.
    # restore_best_weights=True: Eğitim sırasında en iyi doğruluğu veren modelin ağırlıkları korunur.
    EarlyStopping(patience=300, monitor='accuracy', mode='max', restore_best_weights=True),

    # ModelCheckpoint: Modelin en iyi performans gösterdiği noktada kaydedilmesini sağlar.
    # save_best_only=True: Sadece en iyi model kaydedilir.
    ModelCheckpoint("/kaggle/working/diagnosis_GRU_CNN_1.keras", save_best_only=True)
]
```

```python
# Modelin eğitimini başlatıyoruz.
# gru_model.fit() fonksiyonu, modelin eğitim sürecini başlatır ve verilerle modelin nasıl eğitileceğini belirler.

history = gru_model.fit(
    # x_train_gru: Eğitim verilerinin özellikleri. Bu, modelin öğrenmesi için kullanılan veridir.
    x_train_gru,
    
    # y_train_gru: Eğitim verilerinin etiketleri. Modelin doğru tahminleri yapabilmesi için doğru etiketlerle eğitim verilir.
    y_train_gru, 
    
    # batch_size: Her iterasyonda (epoch) kullanılan örnek sayısı. Burada 8 olarak belirlenmiş. 
    # Bu, modelin her seferinde 8 örneği işleyeceği anlamına gelir.
    batch_size=8, 
    
    # epochs: Eğitim verilerinin model üzerinde kaç kez geçirileceğini belirtir. Burada 100 epoch belirlenmiş.
    # Bu, modelin eğitim verileri üzerinde 100 kez çalışacağı anlamına gelir.
    epochs=100, 
    
    # validation_data: Modelin eğitim sırasında doğrulama verisi ile test edilmesi. 
    # Bu veriler, modelin eğitim sırasında nasıl performans gösterdiğini izlemek için kullanılır.
    validation_data=(x_val_gru, y_val_gru),
    
    # callbacks: Eğitim sırasında belirli olayları (callback'leri) tetikleyen fonksiyonlar.
    # Daha önce tanımlanan EarlyStopping ve ModelCheckpoint callback'leri burada kullanılmaktadır.
    callbacks=cb
)
```

```python
# Modelin eğitim verisi üzerinde değerlendirilmesi.
# gru_model.evaluate() fonksiyonu, eğitilmiş modelin bir veri seti (burada eğitim verisi) üzerinde 
# doğruluğunu ve kaybını hesaplar.

gru_model.evaluate(
    # x_train_gru: Değerlendirilmek üzere kullanılan eğitim verilerinin özellikleri.
    x_train_gru,
    
    # y_train_gru: Eğitim verilerinin doğru etiketleri (outputlar).
    y_train_gru
)

# Eğitim sırasında kayıp ve doğruluk eğrilerini çizmek için plot_loss_curves fonksiyonu çağrılıyor.
# Bu fonksiyon, modelin eğitim sürecindeki kayıp ve doğruluk (accuracy) değerlerini çizerek
# eğitim sürecinin nasıl geliştiğini görselleştirir.

plot_loss_curves(history)
```

```python
# Sınıfların isimlerini içeren bir liste tanımlanıyor.
# Bu sınıflar, modelin sınıflandırma yaptığı hastalık türleridir.
classes = ["COPD", "Bronchiolitis", "Pneumonia", "URTI", "Healthy"]

# Test verisi üzerinde modelin tahmin yapması sağlanıyor.
# gru_model.predict() fonksiyonu, test verisi üzerinde modelin tahminlerini yapar.
preds = gru_model.predict(x_test_gru)

# Tahmin edilen sonuçları argmax() fonksiyonu ile en yüksek olasılığa sahip sınıfı seçerek
# tek bir sınıf olarak elde ediyoruz.
# Bu işlem, her bir örnek için modelin hangi sınıfa ait olduğunu belirler.
classpreds = [np.argmax(t) for t in preds]

# Gerçek test sınıfları da benzer şekilde argmax() fonksiyonu ile elde ediliyor.
# Gerçek etiketlerin (y_test_gru) de modelin tahminleriyle karşılaştırılabilmesi için
# aynı şekilde en yüksek olasılığa sahip sınıflar seçiliyor.
y_testclass = [np.argmax(t) for t in y_test_gru]

# Gerçek sınıflar ile tahmin edilen sınıfların karşılaştırılmasını sağlayan 
# karışıklık matrisi (confusion matrix) oluşturuluyor.
cm = confusion_matrix(y_testclass, classpreds)

# Karışıklık matrisini görselleştirmek için bir ısı haritası (heatmap) çiziliyor.
plt.figure(figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')

# Isı haritası çiziliyor. Bu harita, tahminlerin ne kadar doğru olduğunu görsel olarak gösterir.
# 'annot=True' parametresi ile her hücrede sayılar gösteriliyor.
# 'fmt='d'' formatı, sayıları tam sayı olarak yazdırmak için kullanılıyor.
ax = sns.heatmap(cm, cmap='Blues', annot=True, fmt='d', xticklabels=classes, yticklabels=classes)

# Başlık, x ve y etiketleri ekleniyor.
plt.title('')  # Başlık boş bırakılmış.
plt.xlabel('Prediction')  # x ekseni (tahmin edilen sınıflar).
plt.ylabel('Truth')  # y ekseni (gerçek sınıflar).

# Görselleştirme gösteriliyor.
plt.show(ax)
```

```python
# 'classification_report' fonksiyonu, sınıflandırma modelinin performansını değerlendiren
# detaylı metrikleri sağlar. Bu metrikler arasında doğruluk, precision, recall ve f1-score yer alır.
# 'target_names' parametresi, raporda kullanılacak sınıf isimlerini belirtir.

print(classification_report(y_testclass, classpreds, target_names=classes))
```

```python
def gru_diagnosis_prediction(test_audio):
    # Test ses dosyasını yükler. 'librosa.load' fonksiyonu ses dosyasını okur
    # ve ses verisini numpy dizisine dönüştürür. 'sampling_rate' ise örnekleme oranını verir.
    data_x, sampling_rate = librosa.load(test_audio)
    
    # Ses verisini, hızını değiştirecek şekilde geriye doğru esnetir. Bu fonksiyon,
    # sesin hızını değiştirmeden sesin uzunluğunu uzatır.
    data_x = stretch(data_x, 1.2)

    # MFCC (Mel-frequency cepstral coefficients) özelliklerini çıkartır.
    # Bu özellikler, sesin frekans bileşenlerini temsil eder.
    features = np.mean(librosa.feature.mfcc(y=data_x, sr=sampling_rate, n_mfcc=52).T, axis=0)

    # Özellikleri uygun şekle dönüştürür. GRU modeli, 52 özellikli tek bir örnek bekler.
    features = features.reshape(1, 52)

    # GRU modelini kullanarak tahmin yapar. 'np.expand_dims(features, axis=1)' ile
    # verinin boyutu uygun hale getirilir.
    test_pred = gru_model.predict(np.expand_dims(features, axis=1))

    # En yüksek olasılığa sahip sınıfı seçer. 'np.argmax' fonksiyonu, en büyük
    # değeri bulur ve bunun karşılık geldiği sınıfı döndürür.
    classpreds = classes[np.argmax(test_pred[0], axis=1)[0]]

    # Tahminin güvenini hesaplar. Burada 'test_pred.T' ile tahminlerin transpozesini alır,
    # ardından ortalama güven puanını elde eder.
    confidence = test_pred.T[test_pred[0].mean(axis=0).argmax()].mean()

    # Sonuçları yazdırır. Sınıf ve güven oranı ekrana yazdırılır.
    print(classpreds, confidence)
```
